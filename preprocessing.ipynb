```python
plt.plot(df['Date'], df['Close'])
plt.plot(df['Date'], df['Low'])
plt.xlabel('Date (weekly)')
plt.ylabel('Price (close, USD)')
plt.grid(True)
# plt.savefig("amazon.png")
```

Amazon splits:  
- 1998-06-02 2:1
- 1999-01-05 3:1
- 1999-09-05 2:1

```python
filtered_df = df[(df['Date'] > pd.to_datetime("1998-04-26")) & (df['Date'] < pd.to_datetime("1998-12-12"))]

plt.plot(filtered_df['Date'], filtered_df['Close'])
plt.xlabel('Date (weekly)')
plt.ylabel('Price (close, USD)')
plt.xticks(rotation=30)
plt.grid(True)
plt.show()
```

```python
plt.plot(filtered_df['Date'], filtered_df['Volume'])
plt.xlabel('Date (weekly)')
plt.ylabel('Volume')
plt.xticks(rotation=30)
plt.grid(True)
plt.show()
```
